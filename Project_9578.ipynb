Ques 1 which categories have the highest average price and how does the variation in price differ across categories?

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
df =pd.read_csv('products.csv')


sns.boxplot(data=df, x='Category', y='Price')
plt.xlabel('Category')
plt.ylabel('Price')
plt.title('Variation in Price Across Categories')
plt.xticks(rotation=90)
plt.show()
```

Ques 2 Which categories have the highest average product price?

```python
df = pd.read_csv('products.csv')
avg_price = df.groupby('Category')['Price'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=avg_price.index, y=avg_price.values)

plt.title('Average Product Price by Category')
plt.xlabel('Category')
plt.ylabel('Average Price')
plt.xticks(rotation=90)

plt.show()
```

Ques 3 how are product price distributed?

```python
df = pd.read_csv('products.csv')

plt.figure(figsize=(8,5))
sns.histplot(df['Price'], bins=15)

plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

Ques 4 Is there a relationship between Price and Stock?

```python
df = pd.read_csv('products.csv')

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Price', y='Stock')

plt.title('Relationship Between Price and Stock')
plt.xlabel('Price')
plt.ylabel('Stock')

plt.show()
```

Ques 5 Compare average inventory value per product across categories?

```python
df = pd.read_csv('products.csv')
df['Inventory_Value'] = df['Price'] * df['Stock']
avg_inventory = df.groupby('Category')['Inventory_Value'].mean().sort_values(ascending=False)


plt.figure(figsize=(10,5))
sns.barplot(x=avg_inventory.index, y=avg_inventory.values,hue=avg_inventory.index,palette="magma")


plt.title('Average Inventory Value per Product Across Categories')
plt.xlabel('Category')
plt.ylabel('Average Inventory Value')
plt.xticks(rotation=90)
plt.show()
```

Ques 6 Is there a relationship between product availiability and price?

```python
df = pd.read_csv('products.csv')

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Availability', y='Price',color="orange")

plt.title('Relationship Between Product Availability and Price')
plt.xlabel('Availability')
plt.ylabel('Price')

plt.show()
```

Ques 7 what is the correlation between numerical variables such as Price, Stock and Inventory Value?

```python
df = pd.read_csv('products.csv')

df['Inventory_Value'] = df['Price'] * df['Stock']
corr = df[['Price', 'Stock', 'Inventory_Value']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Correlation Between Price, Stock and Inventory Value')
plt.show()
```

Ques 8 Does availiability differ across Top 5 categories

```python
df = pd.read_csv('products.csv')

# Top 5 categories with most products
top_categories = df['Category'].value_counts().head(5).index

filtered_df = df[df['Category'].isin(top_categories)]

plt.figure(figsize=(10,6))
sns.countplot(data=filtered_df, x='Category', hue='Availability',width=0.5)

plt.title('Availability Across Top 5 Categories')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=0)

plt.show()
```

Ques 9 what percentage of products are in stocks, pre-order, or discontinued?

```python
df = pd.read_csv('products.csv')
availability_percent = df['Availability'].value_counts(normalize=True) * 100

plt.figure(figsize=(7,7))
plt.pie(
    availability_percent,
    labels=availability_percent.index,
    autopct='%1.1f%%'
)

plt.title('Percentage of Products by Availability Status')
plt.show()
```

Ques 10 Which color are most common among products?

```python
df = pd.read_csv('products.csv')
top_colors = df['Color'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_colors.index,
            y=top_colors.values,
            hue=top_colors.index,
            palette="pastel"
)

plt.title('Top 10 Most Common Product Colors')
plt.xlabel('Color')
plt.ylabel('Number of Products')
plt.xticks(rotation=25)

plt.show()
```

Ques 11 Compare the distributions of Product Price and Stock Levels using histograms.

```python
df = pd.read_csv('products.csv')
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram of Price
sns.histplot(df['Price'], bins=15, color="purple", ax=axes[0])
axes[0].set_title('Distribution of Product Prices')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Frequency')

# Histogram of Stock
sns.histplot(df['Stock'], bins=15, color="pink", ax=axes[1])
axes[1].set_title('Distribution of Stock Levels')
axes[1].set_xlabel('Stock')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Ques 12 Create a dashboard using four different visualizations to analyze the products dataset:

1. Pie Chart:  Percentage of products by Availability Status.
2. Histogram:  Distribution of Product Prices.
3. Bar Chart:  Top 10 Most Common Product Colors.
4. Scatter Plot:  Relationship between Price and Stock.

```python
df = pd.read_csv('products.csv')

availability = df['Availability'].value_counts()
top_colors = df['Color'].value_counts().head(10)

fig, axes = plt.subplots(2,2, figsize=(15, 10))

# 1. Pie Chart
axes[0,0].pie(
    availability.values,
    labels=availability.index,
    autopct='%1.1f%%'
)
axes[0,0].set_title('Product Availability')

# 2. Histogram
sns.histplot(df['Price'], bins=15, ax=axes[0,1],color="lightgreen")
axes[0,1].set_title('Price Distribution')
axes[0,1].set_xlabel('Price')

# 3. Bar Chart
sns.barplot(
    x=top_colors.index,
    y=top_colors.values,
    hue=top_colors.index,
    palette="pastel",
    legend=False,
    ax=axes[1,0]
)
axes[1,0].set_title('Top 10 Product Colors')
axes[1,0].set_xlabel('Color')
axes[1,0].set_ylabel('Count')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Scatter Plot
sns.scatterplot(
    data=df,
    x='Price',
    y='Stock',
    ax=axes[1,1],
    color="red"
)
axes[1,1].set_title('Price vs Stock')

plt.tight_layout()
plt.show()
```
